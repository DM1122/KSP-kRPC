# Simulator

```python
# external
import numpy as np
import pandas as pd
import plot
from IPython.display import display

# project
from main import models
```

```python
# parameters
x_0 = 0
v_0 = 0
m = 1
b = 0.01
g = 9.81
dt = 0.01
t_max = 10
x_0 = np.array([[1], [1]])

# model instantiation
model = models.ConstantMassModel(x_0=x_0, m=m, b=b, g=g)
```

```python
# preallocation
ts = np.arange(start=0, stop=t_max, step=dt)
xs = np.zeros((2, len(ts)))

xs[:, 0, np.newaxis] = x_0
F_t = np.sin(ts) * 5 + 5

# simulation
model.reset()
for i in range(1, len(ts)):
    x = model(u=F_t[i], dt=dt)

    xs[:, i, np.newaxis] = x

# plotting
dfd = {"t": ts, "F_t": F_t, "x": xs[0, :]}
df = pd.DataFrame(data=dfd)

plot.line(
    df=df, x="t", y=["F_t", "x"], title="Test Line Plot", dark=True, markers=False
)
```

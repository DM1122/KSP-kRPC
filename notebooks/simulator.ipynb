# Simulator

```python
# external
import numpy as np
import pandas as pd
from IPython.display import display
from plot import plot

# project
from main import models
```

```python
# model parameters
x_0 = 0
v_0 = 0
m = 1
b = 0.1
g = 9.81
F_t_max = 10
x_0 = np.array([[1], [1]])

# simulation parameters
dt = 0.01
t_max = 10

# model instantiation
model = models.ConstantMassModel(x_0=x_0, m=m, b=b, g=g, F_t_max=F_t_max)
```

```python
# preallocation
ts = np.arange(start=0, stop=t_max, step=dt)
xs = np.zeros((2, len(ts)))

xs[:, 0, np.newaxis] = x_0
a = np.ones(len(ts))

# simulation
model.reset()
for i in range(1, len(ts)):
    x = model(u=a[i], dt=dt)

    xs[:, i, np.newaxis] = x

# plotting
dfd = {"t": ts, "a": a, "x": xs[0, :]}
df = pd.DataFrame(data=dfd)

plot.line(df=df, x="t", y=["a", "x"], title="Test Line Plot", dark=True, markers=False)
```
